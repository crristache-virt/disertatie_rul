# Exploratory Data Analysis: Backblaze SMART Dataset
Basic statistics, failure imbalance, and SMART feature distributions.

```python
from src.io import load_backblaze_smart
import matplotlib.pyplot as plt
import seaborn as sns

df = load_backblaze_smart()
if df is None:
    print("Dataset missing. Please add Backblaze SMART files to DATA_DIR.")
else:
    print(df.describe())
    print(df["failure"].value_counts())
    sns.countplot(x="failure", data=df)
    plt.title("Failure Imbalance")
    plt.show()
    smart_cols = [c for c in df.columns if c.startswith("smart_")]
    df[smart_cols].hist(figsize=(15, 10))
    plt.suptitle("SMART Feature Distributions")
    plt.show()
```
